```python
import numpy as np
import scipy as sp
from math import pi, sqrt, tanh
import matplotlib.pyplot as plt
import time
from ipywidgets import interact
from os import path
```

```python
Na=10
La=2**Na
Na1=np.arange(0,Na+1)
Na2=Na-Na1
La1= 2**Na1
La2= 2**Na2

Nblist=np.arange(4,10)
plt.figure()
for Nb in Nblist:
    Lb=2**Nb

    p2= (La+Lb)/(La*Lb)
    p3= (La1**2+ La2**2+ Lb**2 + 3*La*Lb)/(La*Lb)**2 

    plt.plot(Na1/Na,p2**2/p3,label='N_B=%d' % Nb)
    # plt.plot(Na1/Na,p2)
    
plt.legend()
plt.show()
```
